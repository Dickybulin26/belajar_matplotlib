### MEMBUAT LEGEND

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. MEMBUAT DATA

```python
def sinusGenerator(amplitudo,frekuensi,t_akhir,theta):
    t = np.arange(0,t_akhir,0.1)
    y = amplitudo * np.sin(2*frekuensi*t + np.deg2rad(theta))
    return t,y

amplitudo = 1
frekuensi = 1
theta = 0
t_akhir = 4

t1,y1 = sinusGenerator(amplitudo,frekuensi,t_akhir,theta)
t2,y2 = sinusGenerator(1,1,4,90)
```

### 2. MEMBUAT PLOT DAN MENAMPILKAN DATA

```python

#* tipe pertama
plt.plot(t1,y1, label='sin(0)')
plt.plot(t2,y2, label='sin(90)')
plt.legend()

plt.grid(True)
plt.show()
```

```python

#* tipe kedua
plt.plot(t1,y1, label='sin(0)')
plt.plot(t2,y2, label='sin(90)')
plt.legend(loc='upper right')

plt.grid(True)
plt.show()
```

```python

#* tipe ketiga
plt.plot(t1,y1, label='sin(0)')
plt.plot(t2,y2, label='sin(90)')
plt.legend(loc='upper center', bbox_to_anchor=(0.5,-0.05))

plt.grid(True)
plt.show()
```

```python

#* tipe keempat
plt.figure(1)
ax = plt.subplot(111)
plt.plot(t1,y1, label='sin(0)')
plt.plot(t2,y2, label='sin(90)')
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width*0.7, box.height])
plt.legend(loc='upper center', bbox_to_anchor=(1.2,1))

plt.grid(True)
plt.show()
```
