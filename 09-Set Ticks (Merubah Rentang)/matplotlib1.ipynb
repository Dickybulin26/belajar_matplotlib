### SET TICKS (MERUBAH RENTANG)

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. MEMBUAT DATA

```python
sudut = np.arange(0,360,1)
y = np.sin(np.deg2rad(sudut))
```

### 2. MEMBUAT PLOT DAN MENAMPILKAN DATA

```python
plt.plot(sudut, y)
plt.grid(True)
plt.ylabel('magnituda')
plt.xlabel('sudut')

plt.yticks([-1, 0, 0.5, 1])
plt.xticks(
    [0, 90, 180, 270, 360],
    [
        r'$ {0}^0 $',
        r'$ {90}^o $',
        r'$ {180}^o $',
        r'$ {270}^o $',
        r'$ {360}^o $',
    ]
)

plt.show()
```
