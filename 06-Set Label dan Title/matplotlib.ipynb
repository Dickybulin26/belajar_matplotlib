### SET LABEL DAN TITLE

```python
import matplotlib.pyplot as plt
import numpy as np
```

### 1. MEMBUAT DATA

```python
def sinusGenerator(amplitudo,frekuensi,t_akhir,theta):
    t = np.arange(0,t_akhir,0.1)
    y = amplitudo * np.sin(2*frekuensi*t + np.deg2rad(theta))
    return t,y

amplitudo = 1
frekuensi = 1
theta = 0
t_akhir = 4

t,y = sinusGenerator(amplitudo,frekuensi,t_akhir,theta)
```

### 2. MEMBUAT PLOT DAN MENAMPILKAN DATA

```python
judul = "Grafik Sinussoidal\n"
rumus = r'$ \mathcal{Y} = A.sin(2 \omega t + \theta) $' + '\n'
parameter1 = r'$ A =  $' + str(amplitudo) + 'cm, '
parameter2 = r'$ \omega =  $' + str(frekuensi) + r'$ \mathit{Hz} $' + ', '
parameter3 = r'$  \theta =  $' + str(theta) + r'$ ^{0} $'

plt.plot(t,y)
plt.title(judul + rumus + parameter1 + parameter2 + parameter3)
plt.xlabel("waktu(detik)")
plt.ylabel("magnituda(cm)")
plt.grid(True)
plt.show()
```
