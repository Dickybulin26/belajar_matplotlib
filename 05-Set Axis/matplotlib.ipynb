### SET AXIS

```python
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Membuat data

```python
def sinusGenerator(amplitudo,frekuensi,t_akhir,theta):
    t = np.arange(0,t_akhir,0.1)
    y = amplitudo * np.sin(2*frekuensi*t + np.deg2rad(theta))
    return t,y
```

```python
t,y = sinusGenerator(1,1,4,0)
```

### Membuat Plot dan Menampilkan Data

```python

#* setting axis, minimum dan maximum
plt.axis([0,4,-2,2])


#* menampilkan data
plt.plot(t,y)
plt.grid(True)
plt.show()
```
