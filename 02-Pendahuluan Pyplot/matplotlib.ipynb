### PENDAHULUAN PYPLOT

```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1. MEMBUAT DATA

```python
x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])

y2 = np.array([1,16,81,256,625])
```

# 2. MEMBUAT PLOT DAN MENAMPILKAN DATA

```python
plt.plot(x,y)
plt.plot(x,y2)
plt.grid(True)
plt.show()
```
