### Instalasi package matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x = [1,2,3,4,5]
# y = [1,2,3,4,5]

PI = np.pi
sudut = np.linspace(0,2*PI,100)
radius = 5

x = radius * np.cos(sudut)
y = radius * np.sin(sudut)
```

inisialisasi  plot

```python
plt.plot(x,y)
plt.grid(True)
plt.show()
```
